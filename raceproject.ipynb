```python
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv("TWO_CENTURIES_OF_UM_RACES.csv")
```

```python
df.head(10)
```

```python
#need records where marathons are of 50km or 50miles and in the year 2020
```

```python
df[(df["Event distance/length"].isin(['50km','50mi'])) & (df["Year of event"]==2020)]
```

```python
df[df["Event name"].str.split("(").str.get(1).str.split(")").str.get(0)=="USA"]
```

```python
#combine all the filters i.e distance,year,country
```

```python
df2=df[(df["Event distance/length"].isin(['50km','50mi'])) & (df["Year of event"]==2020) & (df["Event name"].str.split("(").str.get(1).str.split(")").str.get(0)=="USA")]
df2.head(10)
```

```python
#remove USA label 
```

```python
[df2["Event name"].str.split("(").str.get(0)]
```

```python
df2["Event name"]=df2["Event name"].str.split("(").str.get(0)
```

```python
df2.head()
```

```python
#clean up athelete age
```

```python
df2["athlete_age"]=2020-df2["Athlete year of birth"]
```

```python
df2.head()
```

```python
#remove h from athlete performance
```

```python
df2["Athlete performance"].str.split().str.get(0)
```

```python
#drop columns which are unnecessary
```

```python
df2 = df2.drop(["Athlete club","Athlete country","Athlete year of birth","Athlete age category"],axis=1)
```

```python
df2.head()
```

```python
#clean up null values
```

```python
df2.isna().sum()
```

```python
df2[df2["athlete_age"].isna()== 1]
```

```python
df2.dropna(how="any")
```

```python
#checking for duplicates
df2[df2.duplicated() == True]
```

```python
df2.head(10)
```

```python
#reset index
```

```python
df2.reset_index(drop =True)
```

```python
#fix data types
```

```python
df2.dtypes
```

```python
df2["athlete_age"] = df2["athlete_age"].astype(int)
```

```python
df2["Athlete average speed"] = df2["Athlete average speed"].astype(float)
```

```python
df2.head()
```

```python
#rename columns
```

```python
df3=df2.rename(columns={"Year of event":'year',
                    "Event dates":"race_date",
                    "Event name":'race_name',
                    "Event distance/length":'race_length',
                    "Event number of finishers":'race_finishers'})
```

```python
df3.head()
```

```python
sns.histplot(df3["race_length"])
```

```python
sns.histplot(df3,x='race_length',hue="Athlete gender")
```

```python
sns.displot(df3[df3["race length"]== '50mi']["Athlete average speed"])
```

```python
sns.violinplot(data=df3, x="race_length",y= "Athlete average speed",hue="Athlete gender")
```

```python
#what age groups are best for 50mi (display 20 records)
```

```python
df3.query('race_length == "50mi"').groupby("athlete_age")["Athlete average speed"].agg(['mean','count']).sort_values('athlete_age',ascending=True).query('count>19')
```

```python
#seasons category based on month
```

```python
df3["race_month"]=df3["race_date"].str.split('.').str.get(1)
```

```python
df3["race_season"] = df3["race_month"].apply(
    lambda x: 'Winter' if x in [12, 1, 2] else
              'Spring' if x in [3, 4, 5] else
              'Summer' if x in [6, 7, 8] else
              'Fall'
)
```
